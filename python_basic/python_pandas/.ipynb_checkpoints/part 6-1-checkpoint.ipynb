```python
# titanic 데이터를 로드하여 age, sex, class, alive  만 가져와서
# 1. 10대 미만 여성인 승객의 정보만 추출
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

titanic = sns.load_dataset('titanic')
mask = (titanic["age"] < 10) & (titanic["sex"] == 'female')
pd_titan = titanic.loc[:, ['age', 'sex', 'class', 'alive']]

pd_titan1 = pd_titan.loc[mask,:]
pd_titan1.head()
```

```python
# 2. bins = [1, 15, 25, 35, 60, 99]
#    lables = ['미성년자', '청년', '중년', '장년', '노인']
#    각 라벨별 승객수를 추출
```

```python
pd_titan2 = pd_titan.copy()

pd_titan2['age'].fillna(value = np.round(pd_titan2['age'].mean()),inplace=True)

count, value = np.histogram(pd_titan2['age'], bins=[1, 15, 25, 35, 60, 99])
bins_lables = ['미성년자', '청년', '중년', '장년', '노인']

# 라벨별 추출
for cnt, label in zip(count, bins_lables):
    print(label, ':',cnt)

# pd_titan2['age'] = list(map(int, pd_titan2['age']))
```

```python
# 3. 나이의 평균, 성별 승객수, class별 승객수, 생존자, 사망자별 승객수를 추출
```

```python
pd_titan3 = titanic.copy()
col_list = ['age_avg', '남성', '여성', 'class_count', 'alive_count']
pd_titan3['age_avg'] = pd_titan3['age'].mean()

list_sex = pd_titan3["sex"].value_counts()
pd_titan3["남성"] = list_sex[0]
pd_titan3["여성"] = list_sex[1]

list_alive = pd_titan3["alive"].value_counts()
pd_titan3["dead"] = list_sex[0]
pd_titan3["live"] = list_sex[1]

pd_titan3["pclass"].value_counts()
```

```python
pd_titan3 = titanic.copy()
print('평균 나이 : ',np.round(pd_titan3['age'].mean()))
print()

for(idx, value) in pd_titan3['sex'].value_counts().items():
    print('{} 승객수 : {}'.format(idx, value))
print()
    
for(idx, value) in pd_titan3['alive'].value_counts().items():
    if idx == 'yes':
        print('{} 승객수 : {}'.format('생존자', value))
    else:
        print('{} 승객수 : {}'.format('사망자', value))

```

```python
# 'stock price.xlsx', 'stock valueation.xlsx' 두개의 파일을 로드하여
# 데이터프레임에 저장 한 후 두개의 데이터프레임을 행으로 결합하세요
df_stock_p = pd.read_excel('dataset/stock_price.xlsx')
df_stock_v = pd.read_excel('dataset/stock_valuation.xlsx')
```

```python
행결합 = pd.concat([df_stock_p, df_stock_v], ignore_index=True) # 행단위 결합
행결합.head()
컬럼결합 = pd.concat([df_stock_p, df_stock_v], axis=1) # 컬럼으로 결합
컬럼결합.head()
```

```python
# merge를 사용하여 결합 : 결합하는 키를 id로, id가 같은 값만 추출
merge_i = pd.merge(df_stock_p, df_stock_v, how = 'inner', on='id')
merge_i
```

```python
#양쪽 자료 모두 출력(없는 자료 Nan)
pd.merge(df_stock_p, df_stock_v, how = 'outer', on='id')

 #왼쪽자료 모두 출력(없는 자료 Nan)
pd.merge(df_stock_p, df_stock_v, how = 'left', on='id')

#오른쪽자료 모두 출력(없는 자료 Nan)
pd.merge(df_stock_p, df_stock_v, how = 'right', on='id') 

#왼쪽 키와 오른쪽 키의 컬럼명이 다를 경우
pd.merge(df_stock_p, df_stock_v, how = 'inner', 
         left_on='stock_name', right_on='name') 
```

```python
df_merge = pd.merge(df_stock_p, df_stock_v, how = 'right',
                    left_on='stock_name', right_on='name') 

# df_merge와 df_price를 merge on=None 으로
# df_merge의 id_y -> id로 변경하고
```

```python
df_merge.rename(columns={'id_y' : 'id'}, inplace=True)
pd.merge(df_merge, df_stock_p, on=None)
```

```python
#  on=None 두개의 데이터프레임에 같은 컬럼명의 값이 같으면 추출
print(pd.merge(df_merge.loc[:,['id', 'stock_name']],
              df_stock_p.loc[:, ['id', 'stock_name']], on=None))
print()
print(df_stock_p)
```
