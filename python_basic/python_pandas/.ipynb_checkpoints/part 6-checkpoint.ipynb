```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic = sns.load_dataset('titanic')

# age 와 fare 컬럼만 추출
df = titanic.loc[:,['age','fare']]
df.head()
```

```python
# age 컬럼의 값에 10을 더한 값을 실행
def add_10(n):
    return n+10

# 두 객체의 합을 구하는 함수
def add_two(a,b):
    return a+b

print(add_10(10), add_two(3,4))
```

```python
#age 컬럼의 값에 10을 더한 값을 실행
df['age+10'] = df['age'].apply(add_10)
df.head()

#age 와 fare의 값을 더함
df['age+20'] = df['age'].apply(add_two, b= 20)
df.head()

df['age+30'] = df['age'].apply(lambda x: x + 30)

df.head()
```

```python
df.applymap(add_10) #데이터프레임 모든 원소에 적용
```

```python
#시리즈 객체에 함수 매핑 : 시리즈.apply(함수)
#df.apply(lambda x: x.isnull()) 
df.apply(lambda x: add_two(x['age'], x['fare']), axis=1)

df
```

```python
df.apply(min, axis=1) #행 단위로 함수 적용
df.apply(min) #axis=0 ,열 단뒤로 함수 적용
```

```python
#전체 age의 평균, age, 인원수를 구하세요
avg_age = df.apply(np.mean)['age']
print(avg_age, len(df))
#auto-mpg.csv 파일을 데이터프레임으로 적용,
mpg_df = sns.load_dataset('mpg')
#mpg-hors = mpg + hors로 컬럼을 추가
mpg_df.info()
mpg_df['mpg-hors'] = mpg_df.apply(lambda x :add_two(x['mpg'], 
                                  x['horsepower']), axis=1)
mpg_df.head(5)
```

```python
#데이터프레임 객체에 함수 매핑
#데이터프레임.pipe(함수)
def missing_value(x):
    return x.isnull()

def missing_count(x):
    return missing_value(x).sum()

def total_number_missing(x):
    return missing_count(x).sum()
```

```python
result_value = df.pipe(missing_value)
print(result_value) # 데이터 프레임
print()

result_value = df.pipe(missing_count)
print(result_value) # 시리즈
print()

result_value = df.pipe(total_number_missing)
print(result_value) # 값
```

```python
# 시리즈의 각 원소에 함수를 적용 : sr.apply(함수)
# 데이터프레임의 각 원소에 함수를 적용 : df.applymap(함수)
# 데이터프레임의 행(인덱스)별 컬럼멸 함수 적용
#    df.apply(함수, axis=0) axis=0는 컬럼별 적용 axis=1 행별 적용
# 데이터프레임을 인자로 받는 함수 : df.pipe(함수)
#    -> return 타입에 따라 데이터 타입이 결정 됨
```

```python
#열 분리
stock_df = pd.read_excel('dataset/stock_data.xlsx')
stock_df
# 년, 월, 일을 분리 리스트로 [2018, 07, 02]
# stock_df.연월일 데이터 타입 확인 
print(stock_df['연월일'].dtype)
# datetume으로 되어 있음 -> str로 변경
# 연, 월, 일로 분리 2000-01-01 -> split('-') => [2000,01,01]
stock_df['연월일'] = stock_df['연월일'].astype('str')
print(stock_df['연월일'].dtype)
dates = stock_df['연월일'].str.split('-')
type(dates)

#시리즈의 리스트의 값을 순서대로 가져옴

stock_df['연'] = dates.str.get(0)
stock_df['월'] = dates.str.get(1)
stock_df['일'] = dates.str.get(2)

stock_df.head()
```

```python
# stock-date.csv 파일을 읽어서 연, 월, 일을 구분해서 컬럼으로 추가
stock_csv = pd.read_csv('dataset/stock-data.csv')
stock_csv['Date'] = stock_csv['Date'].astype('str')
csv_date = stock_csv['Date'].str.split('-')

stock_csv['연'] = csv_date.str.get(0)
stock_csv['월'] = csv_date.str.get(1)
stock_csv['일'] = csv_date.str.get(2)
stock_csv.head()
```

```python
# 필터링 : 불린 인덱싱
# 데이터프레임[불린 인덱싱] : True인 행만 가져옴
# 나이가 10대인 자료만 추출

titanic.loc[(10 <= titanic.age) &  (titanic.age< 20),:]


titanic.loc[(10 > titanic.age) &  (titanic.sex == 'female'),:]

titanic.loc[(60 <= titanic.age) &  (titanic.survived == 1),['age', 'sex', 'class', 'alive']]
```

```python
# sibsp 컬럼 : 형제 또는 배우자의 수 -> 수가 3,4,5인 승객의
# age, class, sibsp, alive 컬럼만 추출
col_list = ['age', 'class', 'sibsp', 'alive']

mask = (titanic['sibsp']>=3) & (titanic['sibsp']<=5)
titanic_df = titanic.loc[mask, col_list]
titanic_df.head()
```

```python
# isin() 함수를 이용하여 추출
isin_filter = titanic['sibsp'].isin([3,4,5])
isin_titanic = titanic.loc[isin_filter, col_list]
isin_titanic.head()
titanic
```

```python
# titanic 데이터를 로드해서
# Nan 데이터를 처리 : age는 평균값으로, deck 컬럼은 삭제,
# 다른 컬럼은 이전 값을 대체
# 평균 age보다 적은 age의 승객 중에서 여성과 생존자 만
# age, sex, alive, class 컬럼만 추출
# 가족의 수가 4,5명인 승객의 나이, 가족수, 생존여부를 추출

titanic_ass = titanic.copy()
titanic_ass = titanic_ass.dropna(subset=['deck'], how='any', axis='index')
titanic_ass['age'].fillna(titanic_ass['age'].mean(axis=0), inplace=True)
titanic_ass

titanic_ass.ffill(inplace=True)

titanic_ass_1 = titanic_ass.copy()
```

```python
col_list = ['age', 'sex', 'alive', 'class']
avg = titanic_ass['age'].mean(axis=0)
mask = (avg > titanic_ass['age']) & (titanic_ass['sex'] == 'female') & (titanic_ass['alive'] == 'yes')
titanic_ass = titanic_ass.loc[mask, col_list]
print(avg)
titanic_ass.head()
```

```python
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```

```python
#데이터프레임 합치기 : concat(데이터프레임 리스트)
result1 = pd.concat([df1,df2])
print(result1)

# 기존의 인덱스를 무시하고 인덱스 재설정
result2 = pd.concat([df1,df2], ignore_index=True)
print(result2)

# 열 단위로 합성
result3 = pd.concat([df1,df2], axis=1)
print(result3)

# 인덱스가 같은 행만 조인
result4 = pd.concat([df1,df2], axis=1, join='inner')
print(result4)
```
