```python
# subplot : 하나의 페이지에 여러개의 그래프 그림
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) *np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title("A table of 2 subplot ")
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time')
plt.ylabel('Undamped')
print(ax2)
plt.show()
```

```python
# 선 그래프
# 시도별 진출입 인구수
import pandas as pd

df = pd.read_excel('dataset/시도별 전출입 인구수.xlsx',header=0)
df.head(30)
```

```python
#데이터 전처리 작업

# 1. Nan값 제거

# 전출지의 Nan자료를 수정 -> 이전 자료의 값으로 대체
# Nan자료를 수정 : ffill 이전자료로 대체, bfill : 이후 자료로 대체
# value = 가뵤 -> 지정하는 값으로 수정
df.fillna(method='ffill', inplace=True) 
df.tail(30)
```

```python
# 2. 필요한 자료만 추출

#서울특별시에서 다른 지역으로 전출한 자료만 추출
mask = (df['전출지별'] == '서울특별시' ) & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul
#
print(df_seoul.head())
# 필요없는 자료 삭제 = 전출지별은 같은 자료이므로 삭제
df_seoul.drop('전출지별', axis=1, inplace=True)
print(df_seoul.head())

# 3. 컬럼명 또는 인덱스 재 설정

#전입지별을 진입지로 변경
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
print(df_seoul.head())
#전입지 컬럼을 인덱스로 변경
df_seoul.set_index('전입지', inplace=True)
print(df_seoul.head())
```

```python
df_seoul.T.plot()
```

```python
#서울에서 경기도로 전입한 인구수를 그래프로 확인
sr_one = df_seoul.loc['경기도']
print(sr_one.head())
print(sr_one.index)
print(sr_one.values)
```

```python
plt.figure(figsize=(20,5)) #그래프 사이즈 지정
plt.plot(sr_one.index, sr_one.values, marker='*', markersize=10, color='r')
plt.xticks(size=10, rotation='vertical') #x축의 데이터를 세로로 회전
plt.title("서울에서 경기도로 이전한 인구수") # 그래프의 제목
plt.xlabel('연도', size=20) # x축 제목
plt.ylabel('이동 인구 수', size=20) # y축 제목
plt.legend(labels=["서울 -> 경기도"],loc='best', fontsize=13)
plt.show()
```

```python
# 서울에서 제주도로 이동한 인구수만 검색해서 그래프로 나타내 보세요
sr_jeju = df_seoul.loc['제주특별자치도']
```

```python
plt.figure(figsize=(20,5)) #그래프 사이즈 지정
plt.plot(sr_jeju.index, sr_jeju.values, marker='*', markersize=10, color='r')
plt.xticks(size=10, rotation='vertical') #x축의 데이터를 세로로 회전
plt.title("서울에서 제주특별자치도로 이전한 인구수", size=30) # 그래프의 제목
plt.xlabel('연도', size=20) # x축 제목
plt.ylabel('이동 인구 수', size=20) # y축 제목
plt.legend(labels=["서울 -> 제주특별자치도"],loc='best', fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(20,5)) #그래프 사이즈 지정
plt.plot(sr_one.index, sr_one.values, marker='*', markersize=10, color='r')
plt.xticks(size=10, rotation='vertical') #x축의 데이터를 세로로 회전
plt.title("서울에서 경기도로 이전한 인구수", size=30) # 그래프의 제목
plt.xlabel('연도', size=20) # x축 제목
plt.ylabel('이동 인구 수', size=20) # y축 제목
plt.legend(labels=["서울 -> 경기도"],loc='best', fontsize=13)


# annotate 추가
plt.ylim(50000,800000)
plt.annotate('', 
            xy=(20,620000),
            xytext=(2,290000),
            xycoords='data',
            arrowprops=dict(arrowstyle='->', color="red", lw=3))
plt.annotate('인구 이동 증가 1970 - 1995',
            xy=(10,400000),
            rotation=17,
            va='baseline',
            ha='center',
            fontsize=14)

plt.show()
```

```python
# 서울 -> 경기도  : sr_one
# 서울 -> 제주로 : sr_jeju 이동한 인구수의 그래프를 동시에 표현
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# ax1 에 서울 -> 경기도  : sr_one
ax1.plot(sr_one, marker='o',color='b', markersize=10, label='서울 -> 경기')
ax2.plot(sr_jeju, marker='v',color='r', markersize=10, label='서울 -> 제주')
ax1.set_title('서울 경기 인구 이동')
ax2.set_title('서울 제주 인구 이동')
ax1.legend(loc='best')
ax2.legend(loc='best')

# y축 범위 (최소값, 최대값)
ax1.set_ylim(50000, 800000)
# ax2.set_ylim(50000, 800000)

# x축 연도 라벨 지정 -> 75도 기울여서
ax1.set_xticklabels(sr_one.index, rotation=45)
ax2.set_xticklabels(sr_jeju.index, rotation=45)

# 눈금 라벨 크기
ax1.tick_params(axis='x', labelsize=10)
ax2.tick_params(axis='y', labelsize=10)

plt.show()
```

```python
# 하나의 axes에 여러 개의 그래프를 추가로 그림
# 충청남도, 경상북도, 강원도, 1970 ~ 2017년 까지의 자료만 검색
col_years = list(map(str, range(1970, 2018)))
col_years
df_1 = df_seoul.loc[['충청남도', '경상북도', '강원도'],col_years]
df_1
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years, df_1.loc['충청남도', :], marker='o', markerfacecolor='g',
       color='olive', label='서울->충남')
ax.plot(col_years, df_1.loc['경상북도', :], marker='o', markerfacecolor='b',
       color='skyblue', label='서울->경북')
ax.plot(col_years, df_1.loc['강원도', :], marker='o', markerfacecolor='r',
       color='r', label='서울->강원')
ax.legend(loc='best')
ax.set_title('서울->충남,경북,강원 인구 이동')
ax.set_xlabel('기 간')
ax.set_ylabel('인구 이동 수')
ax.set_xticklabels(col_years, rotation=75)

plt.show()
```

```python
#전라남도, 충청북도, 경상남도, 경기도로 이동한 인구수를
#하나의 화면에 모두 그리세요
#화면을 4개로 분할해서 그래프를 그리세요
```

```python
df_2 = df_seoul.loc[['전라남도', '충청북도', '경상남도', '경기도'],col_years]

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years, df_2.loc['전라남도', :], marker='o', markerfacecolor='g',
       color='olive', label='서울->전남')
ax.plot(col_years, df_2.loc['충청북도', :], marker='o', markerfacecolor='b',
       color='skyblue', label='서울->충북')
ax.plot(col_years, df_2.loc['경상남도', :], marker='o', markerfacecolor='r',
       color='r', label='서울->경남')
ax.plot(col_years, df_2.loc['경기도', :], marker='o', markerfacecolor='r',
       color='r', label='서울->경기도')

ax.legend(loc='best')
ax.set_title('서울->전남,충북,경남,경기도 인구 이동')
ax.set_xlabel('기 간')
ax.set_ylabel('인구 이동 수')
ax.set_xticklabels(col_years, rotation=75)

plt.show()
```

```python
# 서울 -> 경기도  : sr_one
# 서울 -> 제주로 : sr_jeju 이동한 인구수의 그래프를 동시에 표현
sr_1 = df_seoul.loc['전라남도']
sr_2 = df_seoul.loc['충청북도']
sr_3 = df_seoul.loc['경상남도']
sr_4 = df_seoul.loc['경기도']

fig = plt.figure(figsize=(40,20))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# ax1 에 서울 -> 경기도  : sr_one
ax1.plot(sr_1, marker='o',color='b', markersize=10, label='서울 -> 전남')
ax2.plot(sr_2, marker='v',color='r', markersize=10, label='서울 -> 충북')
ax3.plot(sr_3, marker='o',color='g', markersize=10, label='서울 -> 경남')
ax4.plot(sr_4, marker='v',color='g', markersize=10, label='서울 -> 경기도')

ax1.set_title('서울 경기 인구 이동')
ax2.set_title('서울 제주 인구 이동')
ax3.set_title('서울 경기 인구 이동')
ax4.set_title('서울 제주 인구 이동')

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

# x축 연도 라벨 지정 -> 75도 기울여서
ax1.set_xticklabels(sr_one.index, rotation=45)
ax2.set_xticklabels(sr_jeju.index, rotation=45)
ax3.set_xticklabels(sr_one.index, rotation=45)
ax4.set_xticklabels(sr_jeju.index, rotation=45)

# 눈금 라벨 크기
ax1.tick_params(axis='x', labelsize=10)
ax2.tick_params(axis='y', labelsize=10)
ax3.tick_params(axis='x', labelsize=10)
ax4.tick_params(axis='y', labelsize=10)

plt.show()
```
