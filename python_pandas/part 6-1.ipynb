```python
#그룹 연산 : df.groupby(그룹을 만들고자 하는 열 또는 열 리스트)
df = titanic.loc[:,['age','sex','class','fare','survived']]
df.head()
```

```python
df['class'].unique()
len(df)

#class를 기준으로 그룹 생성
class_group = df.groupby(['class'])
list(class_group)
```

```python
#특정 그룹을 선택하고자 할 경우 : 그룹한 결과.get_group(그룹명)
print(class_group.get_group('First').head())
```

```python
# class별 평균 나이
print(class_group.mean())
print()
print(class_group.mean().age)
print()
print(class_group.get_group('First').mean())
print()
print(class_group.get_group('First').mean().age)
print()
print(class_group.get_group('First')['age'].mean())
print()
print(np.floor(class_group.get_group('First')['age'].mean()))
```

```python
# class별 평균 나이와 평균 요금을 출력
class_group.mean().age
class_group.mean().fare
for key, group in class_group:
    print("{}등급의 평균 나이 : {} \t 평균 요금 : {}".format(key,
                                                 np.floor(group.age.mean()),
                                                 np.floor(group.fare.mean())))
# 평균 나이가 가장 많은 클래스의 평균나이를 출력
print("평균 나이가 가장 많은 클래스의 평균 나이 : ",np.floor(class_group.mean().age.max()))
```

```python
# 클래스별 성별 기준으로 그룹
class_sex_group = df.groupby(['class', 'sex'])
list(class_sex_group)
# 가장 나이가 많은 클래스와 성, 적은 클래스와 성을 출력
class_sex_group.mean().age
age_max_list = {}
age_min_list = {}
for key, value in list(class_sex_group.age):
    age_max_list[key] = value.max()
    age_min_list[key] = value.min()

print()
print("max : {} - {}, min : {} - {}".format(max(age_max_list),
                                           age_max_list[max(age_max_list)],
                                           min(age_min_list),
                                           age_min_list[max(age_min_list)]
                                           ))
```

```python
# 그룹 연산 메소드
class_group = df.groupby('class')
std_all = class_group.std()
print(std_all)
```

```python
# fare열의 표준편차가 1등석과 2등석간의 값의 차이가 심함
# fare열만 std() 구함
std_fare = class_group.fare.std()
print(std_fare) # 시리즈로 반환됨
```

```python
# 그룹에 함수를 여러개 적용 : group.agg([함수 리스트])
agg_all = group.agg(['min','max'])
print(agg_all)

# 열 마다 다른 함수를 적용 : group.arr({'열1':함수1, '열2':함수2})
# age는 중간값, 요금은 가장 비싼요금, 가장 싼 요금
agg_sep = class_group.agg({'fare': ['max','min'],
                           'age' : 'mean'})
agg_sep
```

```python
# 그룹 연산 데이터 변환 : group.transforms(매핑함수)
# 그룹별 age 열의 평균 집계 연산
age_mean = class_group.age.mean()
print(age_mean)

# 그룹별 age 열의 표준편차 집계 연산
age_std = class_group.age.std()
print(age_std)
```

```python
# 그룹별 age 열의 iteration으로 z-score를 계산하여 출력
for key, group in class_group.age:
    group_zscore=(class_group - age_mean.loc[key])/age_std.loc[key]
    print(' -- origin : ', key)
    print(group_zscore.head(3))
    print()
```

```python
# ㅋ
def z_score(x):
    return ( x - x.mean()) / x.std()

age_trans = class_group.age.transform(z_score)
print(age_trans.loc[[1,9,0]])
age_trans.head()
```

```python
# 그룹 객체 필터링 : group.filter(조건식 함수)
# 데이터 개수가 200개 이상인 그룹만 데이터프레임으로 변환
filter_group = class_group.filter(lambda x : len(x)>=300)
filter_group.head()
len(filter_group)
```

```python
# 그룹 객체에 함수 매핑  group.apply(매핑함수)
agg_group = class_group.apply(lambda x: x.describe())
agg_group

# age열에 z_score 함수를 적용
age_zscore = class_group.apply(z_score)
age_zscore.head()

# 나이의 평균이 30보다 작은 그룹만 데이터 출력
age_filter = class_group.apply(lambda x: x.age.mean() < 30)
age_filter

for x in age_filter.index:
    if age_filter[x] == True:
        age_filter_df = class_group.get_group(x)
        print(age_filter_df.head())
        print()
```

```python
# 피벗 : pd.pivot_table() : 집계 함수를 행과 열, 데이터 지정
pdf = pd.pivot_table(df, # 피벗할 데이터프레임
                     index = 'class',   # 행으로 갈 컬럼
                     columns='sex',     # 열로 갈 컬럼
                     values='age',      # 데이터로 사용할 컬럼
                     aggfunc=['max', 'min']) # 적용할 그룹 함수
print(pdf.index, pdf.columns)
pdf
```

```python
# 행 인덱스는 : 클래스와 성별, 컬럼 : survived
# 값은 age, fare
# 함수는 mean, max 함수를 적용 pdf1을 생성

pdf1 = pd.pivot_table(df, # 피벗할 데이터프레임
                     index = ['class', 'sex'],   # 행으로 갈 컬럼
                     columns='survived',     # 열로 갈 컬럼
                     values=['age', 'fare'],      # 데이터로 사용할 컬럼
                     aggfunc=['mean', 'max']) # 적용할 그룹 함수
print(pdf1.index, pdf1.columns)
pdf1

# 클래스가 First인 행을 추출
pdf1.xs('First', level='class')
# 클래스가 First이고 female인 행을 추출
pdf1.xs(('First', 'female'))
# sex가 'male'인 행만 추출
pdf1.xs('male', level='sex')
```

```python
# 열을 출력(mean 의 결과만 출력)
pdf1.xs('mean', axis=1)

pdf1.xs(('mean','age'), axis=1)

# survived의 값이 1인 자료만 추출
pdf1.xs(1, level='survived', axis=1)

#컬럼에서 max의 'fare'만 선택해서 추출
pdf1.xs(('max', 'fare',0),level=[0,1,2], axis=1)
```
