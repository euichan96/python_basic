#### 지도학습  - 분류(classification)
- step 01. 데이터 분리(training data, test data) 학습데이터와 테스트데이터
- step 02. 학습데이터를 기반으로 ML 알고리즘을 적용해 학습 모델을 생성
- step 03. 테스트데이터를 기반으로 분류예측을 수행
- step 04. 모델의 성능평가


```python
# 데이터 분류함수
from sklearn.model_selection import train_test_split
from sklearn.tree            import DecisionTreeClassifier
from sklearn.metrics         import accuracy_score
```

```python
print('step 01. \n')
x_train, x_test ,y_train, y_test = train_test_split(iris.data,
                                                    iris.target,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    random_state=100)
#test_size 전체중 20프로를 테스트 데이터로
```

```python
x_train.shape, x_test.shape ,y_train.shape, y_test.shape
```

```python
print('train data - ',x_train)
print('train data - ',y_train)
```

```python
print('train data - ',x_test)
print('train data - ',y_test) # 셔플의 중요성 
```

```python
print('step 02. \n')

iris_DF_model = DecisionTreeClassifier()
iris_DF_model.fit(x_train, y_train) #학습데이터만
```

```python
print('step 03. predict() \n')

y_pred = iris_DF_model.predict(x_test) # 예측결과를 리턴
print('y_test - ',y_test) # 정답
print('y_pred - ',y_pred) # 결과
```

```python
print('step 04. 예측정확도 - \n')

print('acc - ',accuracy_score(y_test, y_pred))
```

```python
print('데이터 프레임 형식에서 학습데이터와 테스트데이터를 분리한다면? - \n')

print('iloc() 이영해서 피처와 타겟을 추출한다면 - ')
iris_feature_FD = iris_DF.iloc[:,0:4]
iris_target_FD = iris_DF.iloc[:, 4]
display(iris_feature_FD)
display(iris_target_FD)
```

```python
iris_DF_model = DecisionTreeClassifier()
iris_DF_model.fit(x_train, y_train)
```

```python
y_pred = iris_DF_model.predict(x_test) # 예측결과를 리턴
print('y_test - ',y_test) # 정답
print('y_pred - ',y_pred, type(y_pred)) # 결과
```

#### 교차검증(croos validation)
- 과적합(overffiting)을 방지하기위한 방법
- 데이터의 편중을 막기위해서
- kFold 방식

```python
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score, cross_validate

# KFold 셔플 없음,   StratifiedKFold 셔플 있음
```

```python
fold_iris = load_iris()

features = fold_iris.data
lable    = fold_iris.target
```

```python
features
```

```python
lable
```

```python
print('5개의 폴더 세트를 분리하여 각 폴더 세트별 정확도를 확인해보자 - ')
cv_acc = []
kfold = KFold(n_splits=5)

fold_dct_model = DecisionTreeClassifier()

for train_idx , test_idx in kfold.split(features):
    #print('train idx - ', train_idx)
    #print('test idx  - ', test_idx)
    x_train, x_val = features[train_idx], features[test_idx]
    y_train, y_val = lable[train_idx], lable[test_idx]
    #print('x_train - ', x_train)
    #print('x_val  - ', x_val)
    fold_dct_model.fit(x_train, y_train)
    
    fold_pred = fold_dct_model.predict(x_val)
    acc = accuracy_score(y_val, fold_pred)
    print('acc - ',acc)
    cv_acc.append(acc)

print('교차검증 평균 정확도 - ', np.mean(cv_acc))
```

```python
print('기존 KFold 방식의 문제점 확인 - \n')

fold_iris_frm = pd.DataFrame(data    = fold_iris.data,
                             columns = fold_iris.feature_names
                            )
fold_iris_frm['target'] = fold_iris.target
fold_iris_frm
```

```python
fold_iris_frm['target'].value_counts()
```

```python
bad_fold_iris = KFold(n_splits = 3)
n_iter = 0

fold_dct_model = DecisionTreeClassifier()
    
for train_idx , test_idx in bad_fold_iris.split(fold_iris.data):
    n_iter+=1
    
    x_train, x_val = fold_iris.data[train_idx], fold_iris.data[test_idx]
    y_train, y_val = fold_iris.target[train_idx], fold_iris.target[test_idx]
    
    lable_train  = fold_iris_frm['target'].iloc[train_idx]
    lable_val    = fold_iris_frm['target'].iloc[test_idx]
    
    #print('교차검증 횟수 - ', n_iter)
    #print('학습 레이블 데이터 분포 - ', lable_train.iloc[:])
    #print('검증 레이블 데이터 분포 - ', lable_val)
    
    fold_dct_model.fit(x_train, y_train)
    
    fold_pred = fold_dct_model.predict(x_val)
    acc = accuracy_score(y_val, fold_pred)
    print('acc - ',acc)
    cv_acc.append(acc)

print('교차검증 평균 정확도 - ', np.mean(cv_acc))
```

#### 연습
- 아이리스 데이터를 이용하여 StratifiedKFold 교차검증을 진행해 보자
- random_state = 200
- StratifiedKFold(3, 5) 평균 정확도 확인

```python
test_iris = load_iris()
test_DF = pd.DataFrame(data = test_iris.data,
                       columns = test_iris.feature_names)
test_DF['target'] = test_iris.target
test_DF
```

```python
cv_acc = []
count=0
St_KFold = StratifiedKFold(n_splits=5)
fold_dct_model = DecisionTreeClassifier(random_state = 200)

for train_idx , test_idx in St_KFold.split(test_iris.data, test_iris.target):
    x_train, x_val = test_iris.data[train_idx], test_iris.data[test_idx]
    y_train, y_val = test_iris.target[train_idx], test_iris.target[test_idx]
    count += 1
    fold_dct_model.fit(x_train, y_train)
    fold_pred = fold_dct_model.predict(x_val)
    acc = accuracy_score(y_val, fold_pred)
    
    print(f'검증횟수{count}, 교차검증 정확도 {acc}')
    cv_acc.append(acc)

print('교차검증 평균 정확도 - ', np.mean(cv_acc))
```
