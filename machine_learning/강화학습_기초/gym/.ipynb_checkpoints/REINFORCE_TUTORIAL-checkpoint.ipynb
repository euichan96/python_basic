# 실습에 필요한 도구
## 1번 도구: Agent와 상호작용 할 수 있는 시뮬레이션 환경 (GYM)

```python
import copy
import gym
import numpy as np
import tensorflow as tf

np.random.seed(12345)
tf.random.set_seed(12345)

env = gym.make('CartPole-v0') # Gym이라는 가상의 놀이터 중, cartpole 놀이기구
env.seed(12345)
env.reset()                   # 가상환경 리셋
```

## 2번 도구: Policy를 학습할 Brain (Neural Network)

```python
# 사용할 Policy용 Nerual Net 구성하기
class Model(tf.keras.Model):
    def __init__(self):
        super(Model, self).__init__(self)
        self.num_actions = env.action_space.n
        self.hidden_layer_0 = tf.keras.layers.Dense(units=128, activation='relu')
        self.hidden_layer_1 = tf.keras.layers.Dense(units=128, activation='relu')
        self.policy_out = tf.keras.layers.Dense(units=self.num_actions, activation='tanh')

    def call(self, inputs):
        inputs = tf.convert_to_tensor(inputs)
        hidden_0 = self.hidden_layer_0(inputs)
        hidden_1 = self.hidden_layer_1(hidden_0)
        policy_val = self.policy_out(hidden_1)
        return policy_val
```

## 3번 도구: REINFORCE 알고리즘

```python
# Gradient들을 업데이트 할 저장창고
class Storage():
    def __init__(self):
        self.total_rewards = []
        self.total_gradients = []

    # 저장소 클리어
    def clear(self):
        self.total_rewards = []
        self.total_gradients = []

    # rewards, gradient값들 저장
    def save(self, reward, gradient):
        gradient = [values.numpy() for values in gradient]
        self.total_rewards.append(reward)
        self.total_gradients.append(gradient)

    # Gt(Return) 계산기
    def discounted_rewards(self, discount_factor):
        discounted_r = copy.deepcopy(np.array(self.total_rewards))
        running_add = 0

        for temp_val in reversed(range(0, len(discounted_r))):
            running_add = np.array(self.total_rewards)[temp_val] + running_add * discount_factor
            discounted_r[temp_val] = running_add

        return discounted_r
```

```python
# 강화 알고리즘
class Brain:
    def __init__(self, model, storage, discount_factor, learning_rate):
        self.model = model
        self.storage = storage
        self.episode = 0
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor
        self.num_actions = env.action_space.n
        self.model.build((1, env.observation_space.shape[0]))
        self.optimizer = tf.optimizers.Adam(self.learning_rate)

    # REINFORCE Cost function
    # -𝑙𝑜𝑔𝜋_𝜃(s,a)
    def cost_fn(self, action_list, probability):
        cost = -tf.reduce_sum(action_list * tf.math.log(probability))
        return cost

    # Action을 확률을 고려해서 뽑아줌
    def action(self, state):
        policy_output = self.model([state])
        temp_action = tf.random.categorical(policy_output, 1)
        action = tf.one_hot(temp_action, self.num_actions)
        probability = tf.nn.softmax(policy_output)
        return action, probability

    # 훈련 시작 함수
    def train(self, env):
        state = env.reset()
        grad_arry = self.model.trainable_weights

        # Grad_arry: Gradient들을 한번에 업데이트 하기 위한 저장창고
        for idx, grad in enumerate(grad_arry):
            grad_arry[idx] = 0

        present_reward = 0
        sum_rewards = 0
        avg_rewards = []
        total_episodes = 450

        while self.episode < total_episodes:

            if self.episode > 400:
                env.render()

            # Gradient를 계산하는 공간
            with tf.GradientTape() as tape:

                # 현재 state에서 action 수행
                action_list, probability = self.action(state)

                # Action에 따라서 환경 이동시킴
                next_state, reward, done, info = env.step(np.argmax(action_list))

                # Cost function 계산
                cost = self.cost_fn(action_list, probability)

            # Tape에 저장되어 있는 cost들에 대해서 gradient를 계산
            grads = tape.gradient(cost, self.model.trainable_weights)

            # rewards와 gradient 저장
            self.storage.save(reward, grads)
            state = next_state

            # 게임 종료
            if done:
                self.episode += 1

                # 1회 시뮬레이션 동안, reward 합
                sum_rewards = np.sum(np.array(self.storage.total_rewards))
                saved_gradient = np.array(self.storage.total_gradients)

                # 1회 시뮬레이션 동안, reward들을 discount해서 하나씩 계산
                saved_discounted_rewards = self.storage.discounted_rewards(self.discount_factor)

                # 𝜃 ← 𝜃+𝛼∇_𝜃 𝑙𝑜𝑔𝜋_𝜃 (𝑠_𝑡,𝑎_𝑡)𝐺_𝑡 수행
                for grads, r in zip(saved_gradient, saved_discounted_rewards):
                    for idx, grad in enumerate(grads):
                        grad_arry[idx] += grad * r / len(saved_gradient)

                self.optimizer.apply_gradients(zip(grad_arry, model.trainable_variables))
                avg_rewards.append(sum_rewards)

                # 최근 10회 마다 reward의 평균을 출력해 줌 (최대 값: 200)
                if self.episode != 0 and self.episode %10 == 0:
                    avg_val = sum(avg_rewards[self.episode-10:self.episode])
                    avg_val = avg_val/10
                    print("가장 최근, 10회 평균 Reward는 {} 입니닷!".format(avg_val))

                # 업데이트 후, grad_arry 초기화
                for idx, grad in enumerate(grad_arry):
                    grad_arry[idx] = 0

                self.storage.clear()
                state = env.reset()
```

## 실제 실행코드

```python
discount_factor = 0.99
learning_rate = 0.001

model = Model()
storage = Storage()
agent = Brain(model, storage, discount_factor, learning_rate)
model.summary()
agent.train(env)
```
