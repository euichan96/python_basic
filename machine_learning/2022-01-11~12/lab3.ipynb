```python
import requests
from bs4 import BeautifulSoup
title = []
info_new = []
urlstr = 'http://media.daum.net/ranking/popular/'
r = requests.get(urlstr)

bs = BeautifulSoup(r.text, 'html.parser')
news = bs.select('ul.list_news2 > li > div > strong > a.link_txt')
info = bs.select('span.info_news')

for news_name in news:
    title.append(news_name.string)
    
for info_name in info:
    info_new.append(info_name.string)

print("-- 뉴스 제목 --")
print(title)
print("-- 뉴스 출판사? 방송사? --")
print(info_new)
with open('output/news.csv', "wt", encoding="utf-8") as f:
    f.write('newstitle, newscomname \n')  
    for i in range(50):
        f.write(title[i]+","+info_new[i]+'\n')  
print(len(title), len(info_new))
```
