```python
import requests
from bs4 import BeautifulSoup
title = []
info_new = []
for n in range(10):
    urlstr = 'https://comic.naver.com/genre/bestChallenge.nhn?page=' + str(n)
    r = requests.get(urlstr)

    bs = BeautifulSoup(r.text, 'html.parser')
    title_n  = bs.select('div.challengeInfo > h6 > a')
    summary_n = bs.select('div.challengeInfo > div.summary')
    rating_n = bs.select('div.challengeInfo > div.rating_type > strong')

    for n in title_n:
        title.append(n.string.strip())
    
    for n in summary_n:
        summary.append(n.string)
        
    for n in rating_n:
        rating.append(n.string)

print(title, summary, rating)

with open('webtoon.csv', "wt", encoding="utf-8") as f:
    f.write('title, summary, rating \n')  
    for i in range(len(title)):
        f.write(title[i]+","+summary[i]+","+rating[i]+'\n')  
print(len(title), len(summary), len(rating))
```
