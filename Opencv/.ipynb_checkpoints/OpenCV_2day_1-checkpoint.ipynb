```python
import numpy as np
import cv2
import sys
```

```python
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)
print(img.shape)

h, w = img.shape[:2]
img_re = cv2.resize(img, (0,0), interpolation = cv2.INTER_AREA)

if img is None:
    print('failed')
    sys.exit()

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)

while True:
    key = cv2.waitKey()
    if key ==27 or key == ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
imgBGR = cv2.imread('./fig/cat.bmp')
imgGRAY = cv2.imread('./fig/cat.bmp', cv2.IMREAD_GRAYSCALE)

if imgBGR is None:
    print('failed')
    sys.exit() 
    
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (20, 6))
plt.subplot(131), plt.imshow(imgBGR), plt.axis('off')
plt.subplot(132), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(133), plt.imshow(imgGRAY, cmap = 'gray'), plt.axis('off')
plt.show()
```

```python
##  

import os

img_list = os.listdir('./fig/images/')
print(type(img_list))

img_files = []
for i in img_list:
    img_file = './fig/images/' + i
    img_files.append(img_file)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN,
                     cv2.WINDOW_FULLSCREEN)
print(img_files)
idx = 0
while True:
    img = cv2.imread(img_files[idx])
    cv2.imshow('image', img)
    
    if cv2.waitKey(3000) == 27:
        break
    
    
cv2.destroyAllWindows
```
